```python
def preprocess_function(examples):
    return {
        "Incorrect": str(examples["Incorrect"]),
        "Corrected": str(examples["Corrected"])
    }
small_dataset = dataset.train_test_split(train_size=0.01, seed=42)["train"]
dataset = small_dataset.map(preprocess_function)
```

```python
from transformers import AutoTokenizer, BartForConditionalGeneration

tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large")

def tokenize_function(examples):
    prefix = "grammar error correction: "
    # Add prefix to incorrect texts
    incorrect_texts = [prefix + str(text) for text in examples["Incorrect"]]
    correct_texts = [str(text) for text in examples["Corrected"]]
    
    # Calculate the increased max_length to accommodate the prefix
    prefix_token_count = len(tokenizer.encode(prefix)) - 2  # -2 for special tokens
    adjusted_max_length = 128 + prefix_token_count
    
    inputs = tokenizer(incorrect_texts, padding="max_length", truncation=True, max_length=adjusted_max_length)
    targets = tokenizer(correct_texts, padding="max_length", truncation=True, max_length=128)

    labels = targets["input_ids"]
    labels = [
        [(token if token != tokenizer.pad_token_id else -100) for token in label_seq]
        for label_seq in labels
    ]
    inputs["labels"] = labels
    return inputs


train_data = train_data.map(tokenize_function, batched=True)
valid_data = valid_data.map(tokenize_function, batched=True)


def correct_grammar(text, model, tokenizer):
    # Add prefix to input text
    PREFIX = "grammar error correction: "
    prefixed_text = PREFIX + text
    
    # Tokenize with adjusted max_length
    prefix_token_count = len(tokenizer.encode(PREFIX)) - 2
    inputs = tokenizer(prefixed_text, return_tensors="pt", padding=True, 
                      truncation=True, max_length=128 + prefix_token_count)
    
    # Move to same device as model
    input_ids = inputs.input_ids.to(model.device)
    attention_mask = inputs.attention_mask.to(model.device)
    
    # Generate output
    outputs = model.generate(
        input_ids=input_ids,
        attention_mask=attention_mask,
        max_length=128,
        num_beams=4,
        early_stopping=True,
        length_penalty=2.0,
        no_repeat_ngram_size=3
    )
    
    # Decode output
    corrected_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    
    return corrected_text
```

```python
import torch
import nltk
from nltk.translate.gleu_score import corpus_gleu
from nltk.tokenize import word_tokenize
from nltk.translate.gleu_score import sentence_gleu
from transformers import EvalPrediction
import numpy as np
import jiwer  # for character error rate (pip install jiwer if not installed)

def postprocess_labels(labels, tokenizer):
    return [
        [token if token != -100 else tokenizer.pad_token_id for token in label_seq]
        for label_seq in labels
    ]
    
def compute_metrics(eval_pred):
    preds = eval_pred.predictions
    labels = eval_pred.label_ids
    preds, labels = eval_pred

    # Convert logits to predicted token ids
    clean_preds = postprocess_labels(preds, tokenizer)
    pred_texts = tokenizer.batch_decode(clean_preds, skip_special_tokens=True)
    clean_labels = postprocess_labels(labels, tokenizer)
    label_texts = tokenizer.batch_decode(clean_labels, skip_special_tokens=True)

    tokenized_preds = [word_tokenize(p,language="english", preserve_line=True) for p in pred_texts]
    tokenized_labels = [word_tokenize(l,language="english", preserve_line=True) for l in label_texts]

    # Compute sentence-level GLEU
    gleu_scores = []
    for pred, ref in zip(tokenized_preds, tokenized_labels):
        if len(pred) >= 1 and len(ref) >= 1:
            try:
                score = sentence_gleu([ref], pred)  # ref must be wrapped in a list
                gleu_scores.append(score)
            except Exception:
                gleu_scores.append(0.0)
        else:
            gleu_scores.append(0.0)

    avg_sentence_gleu = np.mean(gleu_scores)

    # Character Error Rate
    cer_scores = [jiwer.cer(ref, hyp) for ref, hyp in zip(label_texts, pred_texts)]
    cer = np.mean(cer_scores)
    
    # Precision, Recall, F0.5
    TP = 0  # Correct tokens
    FP = 0  # Extra tokens in pred
    FN = 0  # Missing tokens in pred

    for pred, label in zip(tokenized_preds, tokenized_labels):
        min_len = min(len(pred), len(label))
        for i in range(min_len):
            if pred[i] == label[i]:
                TP += 1
            else:
                FP += 1
                FN += 1
        FP += max(0, len(pred) - len(label))  # Remaining pred tokens are FP
        FN += max(0, len(label) - len(pred))  # Remaining label tokens are FN

    precision = TP / (TP + FP) if (TP + FP) > 0 else 0.0
    recall = TP / (TP + FN) if (TP + FN) > 0 else 0.0
    beta_squared = 0.5 ** 2
    f0_5 = (1 + beta_squared) * precision * recall / (beta_squared * precision + recall) if (precision + recall) > 0 else 0.0

    return {
        "gleu": avg_sentence_gleu,
        "cer": cer,
        "precision": precision,
        "recall": recall,
        "f0_5": f0_5
    }
def preprocess_logits_for_metrics(logits, labels):
    if isinstance(logits, tuple):
        logits = logits[0]
    return torch.argmax(logits, dim=-1)
```

```python
from transformers import Seq2SeqTrainer
import torch

class CustomPrefixTrainer(Seq2SeqTrainer):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.prefix = "grammar error correction: "
        
    def prediction_step(
        self,
        model,
        inputs,
        prediction_loss_only,
        ignore_keys=None,
    ):
        """
        Overriding prediction_step to use custom prefix for generation.
        """
        if prediction_loss_only:
            return super().prediction_step(
                model, inputs, prediction_loss_only=True, ignore_keys=ignore_keys
            )
            
        has_labels = "labels" in inputs
        inputs = self._prepare_inputs(inputs)
        
        # Compute loss with labels
        with torch.no_grad():
            loss, generated_tokens = self.compute_loss(model, inputs, return_outputs=True)
            loss = loss.mean().detach()
        
        if self.args.predict_with_generate and not prediction_loss_only:
            # Get tokenizer
            tokenizer = self.tokenizer
            
            # Prepare generation inputs
            generation_inputs = {}
            # Get the input_ids without the prefix
            input_ids = inputs["input_ids"]
            attention_mask = inputs["attention_mask"]
            
            # Add prefix to input_ids
            batch_size = input_ids.shape[0]
            prefix_tokens = tokenizer(
                [self.prefix] * batch_size, 
                return_tensors="pt",
                add_special_tokens=False
            ).input_ids.to(input_ids.device)
            
            # Remove BOS token from input_ids if it exists
            if tokenizer.bos_token_id is not None:
                # Check if input_ids start with BOS token
                if (input_ids[:, 0] == tokenizer.bos_token_id).all():
                    input_ids = input_ids[:, 1:]
                    attention_mask = attention_mask[:, 1:]
            
            # Concatenate prefix with input_ids
            prefixed_input_ids = torch.cat([prefix_tokens, input_ids], dim=1)
            
            # Update attention mask
            prefix_attention_mask = torch.ones(
                (batch_size, prefix_tokens.shape[1]),
                dtype=attention_mask.dtype,
                device=attention_mask.device
            )
            prefixed_attention_mask = torch.cat([prefix_attention_mask, attention_mask], dim=1)
            
            # Set up generation inputs
            generation_inputs["input_ids"] = prefixed_input_ids
            generation_inputs["attention_mask"] = prefixed_attention_mask
            
            # Generate
            generated_tokens = self.model.generate(
                **generation_inputs,
                max_length=128,
                num_beams=4,
                early_stopping=True,
                length_penalty=2.0,
                no_repeat_ngram_size=3
            )
            
            # Get labels
            labels = inputs["labels"]
            
            if has_labels:
                labels = inputs["labels"]
                if labels.shape[-1] < generated_tokens.shape[-1]:
                    labels = self._pad_tensors_to_max_len(labels, generated_tokens.shape[-1])
                elif labels.shape[-1] > generated_tokens.shape[-1]:
                    generated_tokens = self._pad_tensors_to_max_len(
                        generated_tokens, labels.shape[-1]
                    )
            
            return (loss, generated_tokens, labels)
        
        return (loss, None, None)
```

```python
# --- STAGE 1: Freeze most parameters, unfreeze specific parts ---
print("--- Preparing for STAGE 1 ---")

# Freeze all parameters by default
for param in model.parameters():
    param.requires_grad = False

# Unfreeze parameters according to the description:
# 1. The "source encoder" (BART's encoder in our case)
for param in model.model.encoder.parameters():
    param.requires_grad = True
print("Unfroze all BART encoder parameters.")

# 2. BART's positional embeddings
# BART typically has shared token embeddings (model.model.shared)
# and separate positional embeddings for encoder and decoder.
model.model.shared.weight.requires_grad = True # Token embeddings
model.model.encoder.embed_positions.weight.requires_grad = True
# The description is for MT where decoder might be English LM.
# For GEC, decoder also processes "English-like" states.
# It's reasonable to also unfreeze decoder's positional embeddings
# if it helps the decoder part adapt to the encoder's new representations.
# For stricter adherence to "freeze most of BART", you might keep this frozen.
# However, "BART positional embeddings" could refer to both. Let's unfreeze both for now.
model.model.decoder.embed_positions.weight.requires_grad = True
print("Unfroze shared token embeddings and encoder/decoder positional embeddings.")

# 3. The self-attention input projection matrix of BARTâ€™s encoder first layer
# This includes q_proj, k_proj, v_proj, and out_proj of the first encoder layer's self-attention
if hasattr(model.model.encoder, 'layers') and len(model.model.encoder.layers) > 0:
    first_encoder_layer = model.model.encoder.layers[0]
    if hasattr(first_encoder_layer, 'self_attn'):
        for proj_name in ['q_proj', 'k_proj', 'v_proj', 'out_proj']:
            if hasattr(first_encoder_layer.self_attn, proj_name):
                module = getattr(first_encoder_layer.self_attn, proj_name)
                if hasattr(module, 'weight'):
                    module.weight.requires_grad = True
                if hasattr(module, 'bias') and module.bias is not None:
                    module.bias.requires_grad = True
        print("Unfroze self-attention projections of BART's encoder first layer.")

# 4. The lm_head (output layer) must be trainable for loss computation
if hasattr(model, 'lm_head'):
    for param in model.lm_head.parameters():
        param.requires_grad = True
elif model.config.tie_word_embeddings and hasattr(model.model, 'shared'):
    # If lm_head is tied to shared embeddings, and shared is unfrozen, this is covered.
    pass
print("Ensured lm_head is trainable.")

# (Optional but recommended for Stage 1 stability) Unfreeze LayerNorms in the encoder
for layer in model.model.encoder.layers:
    if hasattr(layer, 'self_attn_layer_norm'):
        layer.self_attn_layer_norm.weight.requires_grad = True
        if layer.self_attn_layer_norm.bias is not None:
            layer.self_attn_layer_norm.bias.requires_grad = True
    if hasattr(layer, 'final_layer_norm'):
        layer.final_layer_norm.weight.requires_grad = True
        if layer.final_layer_norm.bias is not None:
            layer.final_layer_norm.bias.requires_grad = True
if hasattr(model.model.encoder, 'layer_norm') and model.model.encoder.layer_norm is not None: # For BART's final encoder LN
     model.model.encoder.layer_norm.weight.requires_grad = True
     if model.model.encoder.layer_norm.bias is not None:
         model.model.encoder.layer_norm.bias.requires_grad = True
print("Unfroze LayerNorms within the encoder.")


# Verify which parameters are trainable (optional)
print("\nTrainable parameters for Stage 1:")
total_params_stage1 = 0
trainable_params_stage1 = 0
for name, param in model.named_parameters():
    total_params_stage1 += param.numel()
    if param.requires_grad:
        # print(name) # Uncomment to list all trainable params
        trainable_params_stage1 += param.numel()
print(f"Total parameters: {total_params_stage1}")
print(f"Trainable parameters in Stage 1: {trainable_params_stage1} ({100 * trainable_params_stage1 / total_params_stage1:.2f}%)")


# Define Training Arguments for Stage 1
# You might want fewer epochs or a different learning rate for this stage.
# The description says the new encoder is trained from bitext, implying a substantial part of training.
# Let's use a moderate number of epochs, e.g., 1/3 to 1/2 of your total planned epochs.
EPOCHS_STAGE_1 = 5 # Example, adjust as needed

training_args_stage1 = Seq2SeqTrainingArguments(
    output_dir="./bart_grammar_correction_stage1",
    eval_strategy="epoch",
    save_strategy="epoch",
    logging_dir="./logs_stage1",
    logging_steps=50, 
    num_train_epochs=EPOCHS_STAGE_1,
    per_device_train_batch_size=16, 
    per_device_eval_batch_size=16, 
    gradient_accumulation_steps=2,   
    fp16=True,                      
    learning_rate=1e-5,             
    weight_decay=0.01,              
    warmup_steps=200,              
    predict_with_generate=True,
    report_to="tensorboard",        
    metric_for_best_model="f0_5",   
    load_best_model_at_end=True,    #loads the best model from this stage for the next
    save_total_limit=2,             # Save fewer checkpoints for stage 1
)

trainer_stage1 = CustomPrefixTrainer( # Or Seq2SeqTrainer if CustomPrefixTrainer is not causing issues
    model=model,
    args=training_args_stage1,
    train_dataset=train_data,
    eval_dataset=valid_data,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)

print(f"\n--- Starting Stage 1 Training for {EPOCHS_STAGE_1} epochs ---")
trainer_stage1.train()
print("--- Stage 1 Training Finished ---")
# The model loaded into 'model' is now the best one from stage 1 due to load_best_model_at_end=True
```
